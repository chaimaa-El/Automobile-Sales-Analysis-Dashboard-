# Data Description

The dataset used for this visualization assignment contains *historical_automobile_sales* data representing automobile sales and related variables during recession and non-recession period.

The dataset includes the following variables:
<br>1. Date: The date of the observation.
<br>2. Recession: A binary variable indicating recession perion; 1 means it was recession, 0 means it was normal.
<br>3. Automobile_Sales: The number of vehicles sold during the period.
<br>4. GDP: The per capita GDP value in USD.
<br>5. Unemployment_Rate: The monthly unemployment rate.
<br>6. Consumer_Confidence: A synthetic index representing consumer confidence, which can impact consumer spending and automobile purchases.
<br>7. Seasonality_Weight: The weight representing the seasonality effect on automobile sales during the period.
<br>8. Price: The average vehicle price during the period.
<br>9. Advertising_Expenditure: The advertising expenditure of the company.
<br>10.Vehicle_Type: The type of vehicles sold; Supperminicar, Smallfamiliycar,                 Mediumfamilycar, Executivecar, Sports.
<br>11.Competition: The measure of competition in the market, such as the number of competitors or market share of major manufacturers.
<br>12.Month: Month of the observation extracted from Date..
<br>13.Year: Year of the observation extracted from Date.
<br>
By examining various factors mentioned above from the dataset, we aim to gain insights into how recessions impacted automobile sales for the company.

# Creating Visualizations for Data Analysis


### We develop a *Line chart* using the functionality of pandas to show how automobile sales fluctuate from year to year


```python
import pandas as pd

# Assuming df is your DataFrame containing the automobile sales data
# Grouping by 'Year' and calculating the average of 'Automobile_Sales'
sales_yearly = df.groupby('Year')['Automobile_Sales'].mean()

# Plotting the line chart
sales_yearly.plot(kind='line', figsize=(10, 6), xlabel='Year', ylabel='Average Automobile Sales', title='Average Automobile Sales Over Years')

# Display the plot
plt.show()
```

### Include the following on the plot
ticks on x- axis with all the years, to identify the years of recession

<br>Title as Automobile Sales during Recession
<br>


```python
import matplotlib.pyplot as plt

# Grouping by 'Year' and calculating the average of 'Automobile_Sales'
sales_yearly = df.groupby('Year')['Automobile_Sales'].mean()

# Plotting the line chart
sales_yearly.plot(kind='line', figsize=(10, 6), xlabel='Year', ylabel='Average Automobile Sales', title='Automobile Sales during Recession')

# Adding ticks on x-axis with all the years
years = range(1980, 2024)
plt.xticks(years, rotation=45)

# Annotating at least two years of recession
plt.text(2008, sales_yearly.loc[2008], 'Recession', ha='center', va='bottom', fontsize=10, color='red')
plt.text(2020, sales_yearly.loc[2020], 'Recession', ha='center', va='bottom', fontsize=10, color='red')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Grouping by 'Year' and calculating the average of 'Automobile_Sales'
sales_yearly = df.groupby('Year')['Automobile_Sales'].mean()

# Plotting the line chart
sales_yearly.plot(kind='line', figsize=(10, 6), xlabel='Year', ylabel='Average Automobile Sales', title='Automobile Sales during Recession')

# Adding ticks for all years
plt.xticks(range(1980, 2024), rotation=45)

# Annotating years of recession
recession_years = df[df['Recession'] == 1]['Year'].unique()  # Get unique recession years
for year in recession_years:
    plt.text(year, sales_yearly.loc[year], 'Recession', ha='center', va='bottom', color='red', fontsize=10)

# Display the plot
plt.show()
```


####  We plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?


```python
import matplotlib.pyplot as plt

# Step 1: Create a separate DataFrame for recession periods
recession_df = df[df['Recession'] == 1]

# Step 2: Group by 'Year' and 'Vehicle_Type', calculate the average of 'Automobile_Sales'
sales_by_vehicle_type = recession_df.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Step 3: Plot the data for each vehicle type
fig, ax = plt.subplots(figsize=(10, 6))

for vehicle_type, data in sales_by_vehicle_type.groupby('Vehicle_Type'):
    ax.plot(data['Year'], data['Automobile_Sales'], label=vehicle_type)

# Step 4: Include labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Average Automobile Sales during Recession')
ax.set_title('Sales Trends by Vehicle Type during Recession')

# Step 5: Display the legend
ax.legend(title='Vehicle Type')

# Display the plot
plt.show()
```

From the plot depicting sales trends by vehicle type during recession periods, several insights can be gleaned:

  Differential Impact of Recession: It appears that different vehicle types experience varying degrees of impact during recession periods. Some vehicle types may exhibit more significant declines in sales, while others may be more resilient or even experience growth.

  Performance of Vehicle Categories: By observing the trends for each vehicle type, we can identify which categories are more susceptible to economic downturns and which ones are more stable. For instance, certain vehicle types may show a consistent downward trend during recessions, indicating weaker demand, while others may show fluctuations or even growth.

  Consumer Preferences: The sales trends provide insights into consumer preferences and purchasing behavior during recessionary periods. Changes in sales volume across different vehicle types may reflect shifts in consumer priorities, such as a preference for more affordable or fuel-efficient vehicles during economic uncertainty.

  Market Response: The market response to recessionary pressures can also be inferred from the plot. For example, certain vehicle types may see a decline in sales due to reduced consumer confidence and spending, while others may benefit from incentives or promotional efforts by manufacturers.

Overall, the plot allows us to analyze the relative performance of various vehicle types during recession periods, providing valuable insights for market analysis, strategic planning, and decision-making within the automotive industry.

### We use the functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group the data by 'Vehicle_Type' and 'Recession', calculate the average of 'Automobile_Sales' for each group
sales_comparison = df.groupby(['Vehicle_Type', 'Recession'])['Automobile_Sales'].mean().reset_index()

# Plot the average sales for each vehicle type during recession and non-recession periods using a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Vehicle_Type', y='Automobile_Sales', hue='Recession', data=sales_comparison)
plt.xlabel('Vehicle Type')
plt.ylabel('Average Automobile Sales')
plt.title('Comparison of Sales Trends per Vehicle Type: Recession vs Non-Recession')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Recession', loc='upper right')
plt.tight_layout()
plt.show()
```

our interpretation captures key insights from the visualization:

  Drastic Decline in Overall Sales: The visualization suggests that there is indeed a significant decrease in the overall sales of automobiles during recession periods. This decline is evident across all vehicle types, as indicated by the lower average sales values during recession compared to non-recession periods.

  Most Affected Vehicle Types: The analysis highlights that certain vehicle types are more severely impacted by recessions than others. Specifically, the executive car and sports categories exhibit the most substantial decline in sales during recession periods. This insight underscores the vulnerability of certain segments of the automotive market to economic downturns.

By identifying the vehicles most affected by recessions, businesses and policymakers can focus their efforts on implementing targeted strategies to mitigate the impact and potentially stimulate demand during challenging economic periods.

### We use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, we want to find more insights from the data to understand the reason. <br> We plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods?
<br>We make use of <u>add_subplot()</u> from Matplotlib for this comparision.


```python
import matplotlib.pyplot as plt

# Filter data for recession and non-recession periods
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]

# Create a figure and subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# Plot GDP variation during recession period
axes[0].plot(recession_data['Year'], recession_data['GDP'], color='red')
axes[0].set_title('GDP Variation During Recession Period')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('GDP')

# Plot GDP variation during non-recession period
axes[1].plot(non_recession_data['Year'], non_recession_data['GDP'], color='blue')
axes[1].set_title('GDP Variation During Non-Recession Period')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('GDP')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

our inference provides a clear interpretation of the plot:

   Impact of Recession on GDP: The plot illustrates that during recession periods, the GDP of the country experienced a notable decline, as indicated by the lower range of GDP values. This observation aligns with economic theory, as recessions typically coincide with reduced economic activity, lower consumer spending, and decreased business investment, all of which contribute to a decline in GDP.

  Connection to Overall Sales: The inference correctly links the observed decline in GDP during recession periods to the potential impact on the overall sales of the company. A lower GDP can signify reduced purchasing power and consumer confidence, which may lead to decreased demand for automobiles and other goods. This insight highlights the interconnectedness between macroeconomic factors, such as GDP, and business performance, emphasizing the importance of understanding broader economic trends for strategic decision-making.

### We develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? We check it for non-recession years to understand the trend

##### We develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>


```python
import matplotlib.pyplot as plt

# Extracting data for non-recession years
non_recession_data = df[df['Recession'] == 0]

# Creating the bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(non_recession_data['Month'], non_recession_data['Automobile_Sales'], s=non_recession_data['Seasonality_Weight']*100, alpha=0.5)

# Adding labels and title
plt.title('Seasonality Impact on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Displaying the plot
plt.show()
```

In this code:

   We first filter the data to include only non-recession years.
  Then, we create a bubble plot using the scatter() function. We pass 'Month' and 'Automobile_Sales' as the x and y coordinates, respectively, and use 'Seasonality_Weight' to represent the size of each bubble (scaled by a factor of 100 for better visualization).
  Finally, we add labels for the axes and a title for the plot, and adjust the x-axis ticks to display month abbreviations.

This bubble plot allows us to visualize the impact of seasonality on automobile sales, showing the variation in sales across different months. The size of each bubble indicates the relative weight of seasonality, providing insights into the seasonal trends in automobile sales.

 ### We create a pie chart to display the portion of advertising expenditure of XYZAutomotives during recession and non-recession periods.
 <br>How did the advertising expenditure of XYZAutomotives change during recession and non-recession periods?


```python
import matplotlib.pyplot as plt

# Create separate DataFrames for recession and non-recession periods
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]

# Calculate the sum of advertising expenditure for both periods
total_advertising_recession = recession_data['Advertising_Expenditure'].sum()
total_advertising_non_recession = non_recession_data['Advertising_Expenditure'].sum()

# Create data for the pie chart
total_advertising = [total_advertising_recession, total_advertising_non_recession]
labels = ['Recession', 'Non-Recession']
colors = ['#ff9999','#66b3ff']

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(total_advertising, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Advertising Expenditure during Recession and Non-Recession Periods')

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

This pie chart will visually represent the portion of advertising expenditure of XYZAutomotives during recession and non-recession periods. Each slice of the pie will correspond to the percentage of total advertising expenditure during each period, with labels indicating whether it is during a recession or non-recession period.



   Higher Advertising Expenditure During Non-Recession Periods: The pie chart illustrates that XYZAutomotives has allocated a larger portion of its advertising expenditure during non-recession periods compared to recession periods. This observation is evident from the larger slice representing non-recession advertising expenditure relative to the recession advertising expenditure.

  Strategic Allocation of Resources: During economic downturns (recession periods), companies often face challenges such as reduced consumer spending and heightened market uncertainty. As a result, they may adjust their marketing strategies and allocate fewer resources to advertising activities. Conversely, during periods of economic stability or growth (non-recession periods), companies may invest more aggressively in advertising to capitalize on consumer confidence and stimulate demand.

By analyzing advertising expenditure patterns across recession and non-recession periods, businesses can gain insights into their strategic resource allocation decisions and adapt their marketing strategies accordingly to navigate through economic cycles effectively.
